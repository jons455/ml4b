We need to tokenise the text data so that we can use ML algorithms on it. Do do so we use the TfidfVectorizer from the scikit-learn package.

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(data["text"].values.astype('U'))
```

# Clustering

## K-means

This produces a sparse matrix which we can use for clustering.

```python
# define arguments for k-means algorithm 
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42, # set random state for reproducability
}
```

K-means requires us to manually set the number of clusters. We will use the elbow rule to identify the optimal number of clusters. We run the algorithm for K equals to 1 or more until 10.

```python
# we create a list which holds sse for each clustering iteration
sse = []
for k in range(1, 11):
    #input the arguments defined in prev cell
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(tfidf)
    sse.append(kmeans.inertia_)
```

We plot the sse for all cluster algorithm runs.

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

There is a evident pivot at k=3. This indicates that this is the optimal number of clusters.

```python
kmeans = KMeans(n_clusters=3, random_state=42).fit(tfidf) # final algortim run with k = 3
```

```python
data["cluster_labels"] = kmeans.labels_ #assign cluster labels to df
```

## Discussion

As already noted HDBScan produces only one (cluster number 158) cluster which appears stable. This cluster contains the words: "find digital twin facebook". This cluster corresponds directly to cluster 1 of k-means in terms of both size and content. The fact that two different algorithms identified the same cluster is evidence of a strong pattern in the data.

The rest of the clusters produced by HDBScan are of diminishing size. These clusters appear to capture the different variations that occur on top of the fist cluster. For example we see that cluster 118 has the words "find digital twin facebook mashable". These are the same words as cluster 158 plus one extra word. Similarly, cluster 139 includes one additional word, "via".

We can see that 2 of the 5 top clusters produced by HDBScan contain the tokens "find digital twin facebook novelty thing youll..." which is indicative of another pattern in the data. Indeed this is confirmed by the K-means results. Cluster 2 of K-means captures all the phrases with the same tokens and some additional ones.

The chief difference between the two algorithms is that K-means groups together all the remaining terms into cluster 0 whilst HDBScan seemingly overfits and tries to create more clusters based on individual word variations. We have attempted to address this issue and discover better structures by running HDBScan a second time with higher values for different parameters. However, the results remain the same which is evidence of cluster stability. In the end the K-means results generalise better.
